```python
import requests
from bs4 import BeautifulSoup

```

# Scraping famous competitive coding questions and DSA Questionnaire Sheet of striver a.k.a takeUforward

## Step 1:- 
### Looking at what the html content actually contains using requests in python?  

```python
url = "https://takeuforward.org/interview-experience/strivers-cp-sheet/?utm_source=youtube&utm_medium=striver&utm_campaign=yt_video"
```

```python
page = requests.get(url)

print(page.text)
```

```python
#!pip install beautifulsoup4
```

## Step 2:- 
### Using beautiful soup to get the soup bowl of maggi and here our maggi is the html content which consists of vegetables like html tag, br tag, link tag and many other tags. So very tasty maggi indead.

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

### Step 2.1:- Getting the detail tag to see what's inside it

```python
results = soup.find(id="details")
```

```python
print(soup.prettify())
```

```python
message = soup
for body in message.find_all('div', {'class': 'body'}):
    if body.find('div', {'class': 'details'}):
        text = body.find('div', {'class': 'details'})
        print(text)
```

## Step 3:-
### Getting all of the links a.k.a 'a tag' from the soup bowl of different tags and store it into array


```python
links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))


print(f"First 10 links are:-\n{links[:10]}\n")
```

## Optional Step or as your homework(Find any python package or any regex sequence matcher that does the check if the url links are valid or not)

```python
import validators
```

```python
valid=validators.url('https://codespeedy.com/')
if valid==True:
    print("Url is valid")
else:
    print("Invalid url")
```

## Step 4:- 
### Finding the pattern in our links that is does the link actually mentions famous competitive programming websites like 'codeforces, spoj, cses' and not a 'profile' one    

```python
storeInCsv=[]
for link in links:
    if ('codeforces' in link or 'spoj' in link  or 'cses' in link) and 'profile' not in link:
        print(link)
        storeInCsv.append(link)
```

## Optional Step manually checking how many links are totally grabbed!

```python
total_link=0
for link in links:
    if ('codeforces' in link or 'spoj' in link  or 'cses' in link) and 'profile' not in link:
        total_link+=1
print(total_link)
```

```python
expected_total=50+50+12+10+11+13+19+1+7+12+22+6+8+35+10+6+12+15+6
expected_total
```

## Last Step:- Saving the list of all of the sites of questions in a csv file. Hurray, now you can see it any time you want to see.

```python
import pandas as pd
df = pd.DataFrame(data={"links": storeInCsv})
df.to_csv("CPstriverCheatSheet.csv", sep=',',index=False)
```
