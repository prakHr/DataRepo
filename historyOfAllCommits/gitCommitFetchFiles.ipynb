```python
path = r"C:\Users\gprak\Downloads\DataRepo"
x = os.listdir(path)
```

```python
def command(file1):
    name,ext = os.path.splitext(file1)
    return(
    f"""
for rev in $(git log --format=%H); do
    git checkout $rev -- {file1}
    cp {file1} ../history/{name}-$rev{ext}
done""")

for file in x:
    if ".py" in file:
        print(command(file))
```
